# Learning OpenCV

```python
import cv2
import os
import matplotlib.pyplot as plt 
from hand_tracker import HandTracker
 
palm_model_path = "./models/palm_detection.tflite"
landmark_model_path = "./models/hand_landmark.tflite"
anchors_path = "./data/anchors.csv"

 

# read image from file
test_img = cv2.imread('./data/test_img.jpg')

# 
detector = HandTracker(palm_model_path, landmark_model_path, anchors_path, box_shift=0.2, box_enlarge=1.3)

kp,box = detector(test_img)

for k in kp:
    cv2.circle(test_img, (int(k[0]), int(k[1])), 50, (0,255,0), thickness=50)

plt.imshow(test_img)
```
